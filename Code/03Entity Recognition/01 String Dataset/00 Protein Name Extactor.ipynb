```python
with open ("protein.info.v11.0.txt", "r", encoding="utf8") as myfile:
    proteins=myfile.readlines()
    myfile.close()
print(type(proteins))
```

```python
print(len(proteins))
```

```python
print(proteins[0])
print(proteins[1])


print(proteins[1].split("	"))
```

```python
protein_names = [protein.split("	")[1] for protein in proteins]
protein_annotations = [protein.split("	")[3] for protein in proteins]

protein_english = [protein.split(";")[0] for protein in protein_annotations]

print(len(proteins))
print(len(protein_names))
print(len(protein_annotations))
```

```python
for p in protein_annotations:
    if "Dystrophin" in p:
        print(p)
```

```python
names = set(protein_english)
print(len(names))


```

```python
from collections import Counter


counted = Counter(protein_english)




print(len(counted))




```

```python
most_common_counted = counted.most_common()


for m in range(len(most_common_counted)):
    print(most_common_counted[m])
    if m>100:
        break
```

```python
for m in range(len(most_common_counted)):
    print(most_common_counted[m][0].split(" "))
    if m>10:
        break
```

```python
most_common_counted = [c[0].lower() for c in most_common_counted if len(c[0].split(" ")) ==1 ]
print(len(most_common_counted))
```

```python
most_common_counted = [m.rstrip() for m in most_common_counted if m != ""]
print(len(most_common_counted))
```
